```
import pandas as pd

df = pd.read_csv('old.csv')
df
```

```
from Rabbit import zg2uni
round1 = lambda x: zg2uni(x)
```

```
df['product_title']= df['product_title'].apply(round1)
df
```

```
import re
def clean_text_round2(text):
    text = re.sub(r'[^\u1000-\u104f]', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```
df['product_title'] = df['product_title'].apply(round2)
df
```

```
pip install bpemb
```

```
from bpemb import BPEmb
bpemb_my = BPEmb(lang="my", vs=200000)
df['product_title']=bpemb_my.encode(df['product_title'])
df
```

```
def clean_text_round3(text):
  underscore="▁"
  text = [''.join(c for c in s if c not in underscore) for s in text]
  return text

round3 = lambda x: clean_text_round3(x)
```

```
df['product_title'] = df['product_title'].apply(round3)
df
```

```
df['category_id'] = df['category'].factorize()[0]
category_id_df = df[['category', 'category_id']].drop_duplicates().sort_values('category_id')
category_to_id = dict(category_id_df.values)
id_to_category = dict(category_id_df[['category_id', 'category']].values)
df
```

```
#stop_words = [line.rstrip('\n') for line in open('stop_word.txt')]
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

def identity_tokenizer(text):
    return text

tfidf = TfidfVectorizer(tokenizer=identity_tokenizer,lowercase=False,sublinear_tf=True,
                        ngram_range=(1,1),min_df=3,max_df=0.2)

features = tfidf.fit_transform(df.product_title).toarray()
labels = df.category_id
features.shape
```

```
from sklearn.feature_selection import chi2
import numpy as np
N = 10
for category, category_id in sorted(category_to_id.items()):
  features_chi2 = chi2(features, labels==category_id)
  indices = np.argsort(features_chi2[0])
  feature_names = np.array(tfidf.get_feature_names())[indices]
  unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
  print("\n '{}':".format(category))
  print("  Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-N:])))
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, indices_train, indices_test = train_test_split (features, labels, df.index,test_size=0.25, random_state = 0)
```

```
from sklearn.svm import LinearSVC

svm = LinearSVC(multi_class='ovr')
svm.fit(X_train, y_train)
svm_pred=svm.predict(X_test)
```

```
from sklearn import metrics
print(metrics.classification_report(y_test, svm_pred))
```

```
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_test, svm_pred)
fig, ax = plt.subplots(figsize=(6,5))
sns.heatmap(conf_mat, annot=True, fmt='d',
            xticklabels=category_id_df.category_id.values, yticklabels=category_id_df.category_id.values)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```
from IPython.display import display
for predicted in category_id_df.category_id:
  for actual in category_id_df.category_id:
    if predicted != actual and conf_mat[actual, predicted] >= 10:
      print("'{}' predicted as '{}' : {} examples.".format(id_to_category[actual], id_to_category[predicted], conf_mat[actual, predicted]))
      display(df.loc[indices_test[(y_test == actual) & (svm_pred == predicted)]][['category', 'product_title']])
      print('')
```

```
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
lg.fit(X_train,y_train)
lg_pred=lg.predict(X_test)
```

```
from sklearn import metrics
print(metrics.classification_report(y_test, lg_pred))
```

```
conf_mat = confusion_matrix(y_test, lg_pred)
fig, ax = plt.subplots(figsize=(6,5))
sns.heatmap(conf_mat, annot=True, fmt='d',
            xticklabels=category_id_df.category_id.values, yticklabels=category_id_df.category_id.values)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```
for predicted in category_id_df.category_id:
  for actual in category_id_df.category_id:
    if predicted != actual and conf_mat[actual, predicted] >= 10:
      print("'{}' predicted as '{}' : {} examples.".format(id_to_category[actual], id_to_category[predicted], conf_mat[actual, predicted]))
      display(df.loc[indices_test[(y_test == actual) & (lg_pred == predicted)]][['category', 'product_title']])
      print('')
```

```
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_pred=rf.predict(X_test)
```

```
from sklearn import metrics
print(metrics.classification_report(y_test, rf_pred))
```

```
rf_conf_mat = confusion_matrix(y_test, rf_pred)
fig, ax = plt.subplots(figsize=(6,5))
sns.heatmap(rf_conf_mat, annot=True, fmt='d',
            xticklabels=category_id_df.category_id.values, yticklabels=category_id_df.category_id.values)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```
for predicted in category_id_df.category_id:
  for actual in category_id_df.category_id:
    if predicted != actual and rf_conf_mat[actual, predicted] >= 10:
      print("'{}' predicted as '{}' : {} examples.".format(id_to_category[actual], id_to_category[predicted], rf_conf_mat[actual, predicted]))
      display(df.loc[indices_test[(y_test == actual) & (rf_pred == predicted)]][['category', 'product_title']])
      print('')
```
